## CNN 모델 생성

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# 이미지 생성
image = tf.constant([[[[1], [2], [3]],
                     [[4], [5], [6]],
                     [[7], [8], [9]]
                    ]], dtype = np.float32)

# 이미지 출력
print(image.shape) # (batch, height, width, channel)
plt.imshow(image.numpy().reshape(3, 3), cmap = 'Greys')
plt.show()

# 패딩 없이 convolution

print('image.shape', image.shape)
weight = np.array([[[[1.]], [[1.]]],
                   [[[1.]], [[1.]]]])
print('weight.shape', weight.shape) # (height, width, channel, batch)

weight_init = tf.constant_initializer(weight)

# 패딩을 주지 않고 연산 -> 크기가 작아진다.
conv2d = keras.layers.Conv2D(filters=1, kernel_size=2, padding='VALID',
                             kernel_initializer=weight_init)(image)

print('conv2d.shape', conv2d.shape) # (batch, height, width, channel)
print(conv2d.numpy().reshape(2,2)) # convolution 결과
plt.imshow(conv2d.numpy().reshape(2,2), cmap='gray')
plt.show()
```
